### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
import pandas as pd
df_loja1 = pd.DataFrame(loja)
df_loja2 = pd.DataFrame(loja2)
df_loja3 = pd.DataFrame(loja3)
df_loja4 = pd.DataFrame(loja4)
```

```python
faturamento_loja1 = df_loja1['Preço'].sum()
faturamento_loja2 = df_loja2['Preço'].sum()
faturamento_loja3 = df_loja3['Preço'].sum()
faturamento_loja4 = df_loja4['Preço'].sum()

print(f"Faturamento Loja 1: R${faturamento_loja1:,.2f}")
print(f"Faturamento Loja 2: R${faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R${faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R${faturamento_loja4:,.2f}")
```

# 2. Vendas por Categoria

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

medias = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(9, 5))

plt.plot(lojas, medias, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8)

plt.title('Média de Avaliação dos Clientes por Loja (Gráfico de Linha)')
plt.xlabel('Loja')
plt.ylabel('Média de Avaliação')

plt.ylim(0, 5)

for i, media in enumerate(medias):
    plt.text(i, media + 0.1, f'{media:.2f}', ha='center', va='bottom', fontsize=12)

plt.grid(True, axis='y', linestyle='--')

plt.show()
```
